```python
bike_rental_data["datetime"] = pd.to_datetime(bike_rental_data["datetime"])
df1 = bike_rental_data.set_index("datetime")
hourly1 = (
    df1.groupby("location_id").resample("1h").size().reset_index(name="count")
)
hourly1.head()
```

```python
hourly1.tail()
```

```python
direct_pick_up_data["datetime"] = pd.to_datetime(
    direct_pick_up_data["datetime"]
)
df2 = direct_pick_up_data.set_index("datetime")
hourly2 = (
    df2.groupby("location_id").resample("1h").size().reset_index(name="count")
)
hourly2.head()
```

```python
hourly2.tail()
```

```python
merged_hourly = hourly1.merge(
    hourly2,
    on=["datetime", "location_id"],
    how="outer",
    suffixes=("_rentals", "_pickups"),
)
```

```python
merged_hourly[merged_hourly["location_id"] == 2].head()
```

```python
merged_hourly.fillna(0, inplace=True)
merged_hourly.head()
```

```python
merged_hourly["total_count"] = (
    merged_hourly["count_rentals"] + merged_hourly["count_pickups"]
)
merged_hourly.head()
```

```python
# Pandas use ISO weekday where Monday is 0 and Sunday is 6
merged_hourly["weekday"] = merged_hourly["datetime"].dt.weekday
merged_hourly.weekday.value_counts()
```

```python
merged_hourly.tail()
```

```python
merged_hourly["is_weekend"] = merged_hourly["weekday"].isin([5, 6]).astype(int)
merged_hourly.head()
```

```python
merged_hourly["year"] = merged_hourly["datetime"].dt.year
merged_hourly["month"] = merged_hourly["datetime"].dt.month
merged_hourly["day"] = merged_hourly["datetime"].dt.day
merged_hourly["date"] = merged_hourly["datetime"].dt.date
merged_hourly["hour"] = merged_hourly["datetime"].dt.hour
merged_hourly["quarter"] = merged_hourly["datetime"].dt.quarter
merged_hourly["is_month_start"] = merged_hourly[
    "datetime"
].dt.is_month_start.astype(int)
merged_hourly["is_month_end"] = merged_hourly[
    "datetime"
].dt.is_month_end.astype(int)
merged_hourly["time_of_day"] = pd.cut(
    merged_hourly["hour"],
    bins=[0, 6, 12, 18, 24],
    labels=["night", "morning", "afternoon", "evening"],
    right=False,
)
```

```python
merged_hourly.head()
```

```python
merged_hourly["date"] = pd.to_datetime(merged_hourly["date"])
```

```python
merged_hourly = pd.get_dummies(
    merged_hourly, columns=["time_of_day"], dtype=int
)
merged_hourly.head()
```

```python
weather_data = data_import_csv(constants.F_WEATHER)
weather_data.drop(columns=["id"], inplace=True)
weather_data["datetime"] = pd.to_datetime(weather_data["datetime"])
weather_data = pd.get_dummies(weather_data, columns=["conditions"], dtype=int)
merged_data = data_merger(
    merged_hourly,
    weather_data,
    on_cols=["datetime"],
    how_to="left",
    suffixe_str=("", "_weather"),
)
```

```python
merged_data.shape
```

```python
holiday_data = data_import_csv(constants.F_HOLIDAYS)
holiday_data.drop(columns=["id"], inplace=True)
holiday_data["date"] = pd.to_datetime(holiday_data["date"])
holiday_data.head()
```

```python
merged_with_holiday = data_merger(
    merged_data, holiday_data, on_cols=["date"], how_to="left", suffixe_str=()
)
merged_with_holiday.head()
```

```python
merged_with_holiday["holiday"].value_counts().sum()
```

```python
merged_with_holiday["is_holiday"] = (
    merged_with_holiday["holiday"].notna().astype(int)
)
merged_with_holiday.head()
```

```python
merged_with_holiday["is_holiday"].value_counts()
```

```python
merged_with_holiday.columns
```

```python
merged_with_holiday["is_weekday_holiday"] = (
    (merged_with_holiday["is_weekend"] == 0)
    & (merged_with_holiday["is_holiday"] == 1)
).astype(int)
merged_with_holiday.head()
```

```python
df = merged_with_holiday.copy()
```

```python
df["holiday_impact"] = df.groupby("holiday")["total_count"].transform("mean")
```

```python
df.holiday_impact.value_counts()
```

```python
df["deviation_from_normal"] = df["total_count"] - df["holiday_impact"]
```
